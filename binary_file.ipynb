```python
# importing library
import pickle
import os

# making menu-driven function.
def add():
    f = open("book.dat", "ab")
    b_code = int(input("Enter book code: "))
    b_name = input("Book name: ")
    aut = input("Enter author name: ")
    price = int(input("Enter price of the book: "))
    L = [b_code, b_name, aut, price]
    
    pickle.dump(L, f)
    print("Data saved")
    f.close()

def delete():
    f1 = open("book.dat", "rb")
    f2 = open("temp.dat", "wb")
    flag = 0
    b_code = int(input("Enter book code to delete: "))
    
    try:
        while True:
            x = pickle.load(f1)
            if x[0] != b_code:
                pickle.dump(x, f2)
            else:
                flag = 1
    except EOFError:
        f1.close()
        f2.close()
    
    os.remove("book.dat")
    os.rename("temp.dat", "book.dat")
    
    if flag == 0:
        print("Sorry, given number not found")
    else:
        print("Deleted")
        
def update():
    f1 = open("book.dat", "rb")
    f2 = open("temp.dat", "wb")
    flag = 0
    b_code = int(input("Enter book code to update: "))
    
    try:
        while True:
            x = pickle.load(f1)
            if x[0] == b_code:
                au = input("Enter new author name: ")
                p = int(input("Enter new price: "))
                L = [x[0], x[1], au, p]
                pickle.dump(L, f2)
                flag = 1
            else:
                pickle.dump(x, f2)
    except EOFError:
        f1.close()
        f2.close()
        
        os.remove("book.dat")
        os.rename("temp.dat", "book.dat")
        
    if flag == 0:
        print("Sorry! Given number not found")
        
def display():
    print("================================")
    print("B CODE\tB NAME\tAUTHOR\tPRICE")
    print("================================")
    
    f = open("book.dat", "rb")
    
    try:
        while True:
            x = pickle.load(f)
            for i in x:
                print(i, end="\t")
            print()
    except EOFError:
        f.close()
        
def search():
    f = open("book.dat", "rb")
    b_name = input("Enter book name to search: ")
    flag = 0
    
    try:
        while True:
            x = pickle.load(f)
            if x[1].lower() == b_name.lower():
                flag = 1
                for i in x:
                    print(i, end="\t")
                print()
    except EOFError:
        f.close()
        
    if flag == 0:
        print("Sorry, given topic not found")
        

    
def cost_search():
    f = open("book.dat", "rb")
    flag = 0
    
    try:
        while True:
            x = pickle.load(f)
            if x[3] >= 500:
                flag = 1
                for i in x:
                    print(i, end="\t")
                print()
    except EOFError:
        f.close()
                
# menu-driven option.
print("1. ADD\n2. Delete\n3. Update\n4. Display\n5. Search\n6. Cost_Search")
ch = "y"
while ch.lower() == "y":
    op = int(input("Enter option to start: "))
    
    if op == 1:
        add()
    elif op == 2:
        delete()
    elif op == 3:
        update()
    elif op == 4:
        display()
    elif op == 5:
        search()
    elif op == 6:
        cost_search()
    else:
        print("Invalid option selected. Please try again")
        
    ch = input("Do you want to continue? (y/n)")
```
